```python
with open("medical_pharmacology_corpus.txt", "r", encoding="utf-8") as f:
    text = f.read()

print("Dataset Loaded Successfully!")
print("Total Characters:", len(text))
print("Total Words:", len(text.split()))
```

```python
print(text[:1000])
```

```python
model1 = markovify.Text(text, state_size=1)
model2 = markovify.Text(text, state_size=2)
model3 = markovify.Text(text, state_size=3)
print("Models created successfully!")
```

```python
print("STATE SIZE = 1\n")

for i in range(5):
    print(model1.make_sentence())
    print()
```

```python
print("STATE SIZE = 2\n")

for i in range(5):
    print(model2.make_sentence())
    print()
```

```python
print("STATE SIZE = 3\n")

for i in range(5):
    print(model3.make_sentence())
    print()
```

```python
print("SHORT MEDICAL STATEMENTS\n")

for i in range(5):
    print(model2.make_short_sentence(120))
    print()
```
